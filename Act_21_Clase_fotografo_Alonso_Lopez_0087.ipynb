```python
# Definición de la Clase Fotografo
class Fotografo:
    # Método constructor para inicializar los atributos
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Tomar Foto
    def tomar_foto(self):
        print(f"📸 {self.nombre} está ajustando el lente y capturando una ráfaga de fotos.")

    # Función 2: Editar Imagen
    def editar_imagen(self):
        print(f"💻 {self.nombre} está en la computadora ajustando el brillo y el contraste.")

    # Función 3: Entregar Álbum
    def entregar_album(self):
        print(f"📁 {self.nombre} ha enviado el enlace con las fotos editadas al cliente.")

# --- Creación de Objetos (Instancias) ---

# Objeto 1: Fotografo1
fotografo1 = Fotografo("Carlos Pérez Fotografo 1 ", "555-0123", 28)

# Objeto 2: Fotografo2
fotografo2 = Fotografo("Elena García Fotografo 2 ", "555-9876", 34)

# --- Ejecución de Funciones y Mensajes ---

print(f"--- Ficha de: {fotografo1.nombre} (Edad: {fotografo1.edad}) (telefono: {fotografo1.telefono}) ---")
fotografo1.tomar_foto()
fotografo1.editar_imagen()
fotografo1.entregar_album()

print("\n" + "-"*90 + "\n")

print(f"--- Ficha de: {fotografo2.nombre} (Edad: {fotografo2.edad}) (telefono: {fotografo2.telefono}) ---")
fotografo2.tomar_foto()
fotografo2.editar_imagen()
fotografo2.entregar_album()

print("\n" + "-"*90 + "\n")
print("Creador de software Alonso Lopez NL = 31")
```
